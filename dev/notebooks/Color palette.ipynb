```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('paper')
```

```python
def show_style(style='default', N_colors=6):
    with plt.style.context(style):
        np.random.seed(0)
        fig, axes = plt.subplots(2)
        ax0 = axes[0]
        ax1 = axes[1]
        for x in range(N_colors):
            ax0.plot(np.linspace(0, 2*np.pi, 100),
                     np.sin(np.linspace(0, 3*np.pi, 100))+x)
        ax1.axvline(np.pi/3, 0, 1, ls='--')
        ax1.axvline(5*np.pi/3, 0, 1, ls='--')
        ax1.axvline(np.pi, 0, 1, label='Default line color')
        for i in range(3):
            if i == 0:
                lab = 'Histogram'
            else:
                lab = None
            ax1.hist(np.random.normal(np.pi, np.pi/(i+1), 10000),
                     bins=np.arange(0,6.5,0.25),
                     density=True,
                     label=lab)
        ax1.legend()

        ax1.set_xlabel('x')
        ax1.set_ylabel(r'$\mathrm{CDF}(x)$')
        ax0.set_ylabel(r'$y=f(x)$')
        ax0.set_title(style+' style')
        plt.show()
```

```python
show_style('default', N_colors=10)
```

```python
show_style('kraken', N_colors=13)
```

```python
show_style('zhunter0', N_colors=12)
```

```python
show_style('seaborn-v0_8-colorblind', N_colors=6)
```
